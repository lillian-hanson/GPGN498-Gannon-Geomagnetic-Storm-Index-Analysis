```python
import json
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
from scipy.signal import welch
```

## Part 1

```python
# Load the JSON data
with open("EYR20230101.json", "r") as f:
    raw_data = json.load(f)

# Extract relevant fields
times = [datetime.fromisoformat(t.replace("Z", "")) for t in raw_data["datetime"]]  # Convert datetime strings to datetime objects
X = np.array(raw_data["X"], dtype=float)
Y = np.array(raw_data["Y"], dtype=float)
Z = np.array(raw_data["Z"], dtype=float)

# Convert to pandas DataFrame for easier handling
df_EYR = pd.DataFrame({
    'time': times,
    'X': X,
    'Y': Y,
    'Z': Z
})

# Set the time as the index
df_EYR.set_index('time', inplace=True)

# Interpolate missing (NaN) values using linear interpolation
df_EYR_inter = df_EYR.interpolate(method='linear', axis=0)

# Calculate horizontal field strength (H) for both raw and interpolated data
H_EYR_raw = np.sqrt(X**2 + Y**2)
H_EYR_interpolated = np.sqrt(df_EYR_inter['X']**2 + df_EYR_inter['Y']**2)

# Plot the raw vs. interpolated data
plt.figure(figsize=(14, 8))

# Raw Horizontal field
plt.subplot(2, 1, 1)
plt.plot(times, H_EYR_raw, label='Raw Horizontal Field (H)', color='blue')
plt.plot(df_EYR_inter.index, H_EYR_interpolated, label='Interpolated Horizontal Field (H)', color='orange', linestyle='--', linewidth=0.5)
plt.ylabel('nT')
plt.title('Horizontal Magnetic Field (H) - Raw vs Interpolated')
plt.xlim(datetime(2023, 1, 1), datetime(2024, 1, 1))
plt.legend()
plt.grid(True)

# Raw Vertical field
plt.subplot(2, 1, 2)
plt.plot(times, Z, label='Raw Vertical Field (Z)', color='red')
plt.plot(df_EYR_inter.index, df_EYR_inter['Z'], label='Interpolated Vertical Field (Z)', color='yellow', linestyle='--', linewidth=0.5)
plt.ylabel('nT')
plt.title('Vertical Magnetic Field (Z) - Raw vs Interpolated')
plt.xlabel('Date')
plt.xlim(datetime(2023, 1, 1), datetime(2024, 1, 1))
plt.legend()
plt.grid(True)

plt.suptitle("Magnetic Fields at EYR (Low Latitude)", fontsize=17)
plt.tight_layout()
plt.show()
```

```python
# Load the JSON data
with open("JCO20230101.json", "r") as f:
    raw_data = json.load(f)

# Extract relevant fields
times = [datetime.fromisoformat(t.replace("Z", "")) for t in raw_data["datetime"]]  # Convert datetime strings to datetime objects
X = np.array(raw_data["X"], dtype=float)
Y = np.array(raw_data["Y"], dtype=float)
Z = np.array(raw_data["Z"], dtype=float)

# Convert to pandas DataFrame for easier handling
df_JCO = pd.DataFrame({
    'time': times,
    'X': X,
    'Y': Y,
    'Z': Z
})

# Set the time as the index
df_JCO.set_index('time', inplace=True)

# Interpolate missing (NaN) values using linear interpolation
df_JCO_inter = df_JCO.interpolate(method='linear', axis=0)

# Calculate horizontal field strength (H) for both raw and interpolated data
H_JCO_raw = np.sqrt(X**2 + Y**2)
H_JCO_interpolated = np.sqrt(df_JCO_inter['X']**2 + df_JCO_inter['Y']**2)

# Plot the raw vs. interpolated data
plt.figure(figsize=(14, 8))

# Raw Horizontal field
plt.subplot(2, 1, 1)
plt.plot(times, H_JCO_raw, label='Raw Horizontal Field (H)', color='blue')
plt.plot(df_JCO_inter.index, H_JCO_interpolated, label='Interpolated Horizontal Field (H)', color='orange', linestyle='--', linewidth=0.5)
plt.ylabel('nT')
plt.title('Horizontal Magnetic Field (H) - Raw vs Interpolated')
plt.xlim(datetime(2023, 1, 1), datetime(2024, 1, 1))
plt.legend()
plt.grid(True)

# Raw Vertical field
plt.subplot(2, 1, 2)
plt.plot(times, Z, label='Raw Vertical Field (Z)', color='red')
plt.plot(df_JCO_inter.index, df_JCO_inter['Z'], label='Interpolated Vertical Field (Z)', color='yellow', linestyle='--', linewidth=0.5)
plt.ylabel('nT')
plt.title('Vertical Magnetic Field (Z) - Raw vs Interpolated')
plt.xlabel('Date')
plt.xlim(datetime(2023, 1, 1), datetime(2024, 1, 1))
plt.legend()
plt.grid(True)

plt.suptitle("Magnetic Fields at JCO (High Latitude)", fontsize=17)
plt.tight_layout()
plt.show()
```

```python
df_JCO_inter['H'] = np.sqrt(df_JCO_inter['X']**2 + df_JCO_inter['Y']**2)

# Convert datetime to elapsed time in days
time_days = (df_JCO_inter.index - df_JCO_inter.index[0]).total_seconds() / (24*3600)

# Function to detrend using linear fit
def detrend(data, time):
    coeffs = np.polyfit(time, data, 1)  # Linear fit
    trend = np.polyval(coeffs, time)
    return data - trend

# Detrend Z and H
Z_detrended = detrend(df_JCO_inter['Z'].values, time_days)
H_detrended = detrend(df_JCO_inter['H'].values, time_days)

# Compute power spectra
fs = 1 / 60  # Sampling frequency: 1 sample per minute = 1/60 Hz
f_z, Pz = welch(Z_detrended, fs=fs, nperseg=4096)
f_h, Ph = welch(H_detrended, fs=fs, nperseg=4096)

# Plot the power spectra
plt.figure(figsize=(12, 6))

plt.loglog(f_z, Pz, label='Z Detrended')
plt.loglog(f_h, Ph, label='H Detrended')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power Spectral Density')
plt.title('Power Spectra of Z and H (Detrended), JCO')
plt.legend()
plt.grid(True, which='both', ls='--')
plt.tight_layout()
plt.show()
```

## Part 2

```python
from urllib.request import urlopen
from io import StringIO
import re

# Download data
url = "https://wdc.kugi.kyoto-u.ac.jp///dstae///wwwtmp/WWW_dstae00581378.dat"
with urlopen(url) as response:
    lines = [line.decode('utf-8').strip() for line in response]

# Keep only rows where the first token is a date
start_idx = next(i for i, line in enumerate(lines) if line.startswith("DATE"))
data_lines = lines[start_idx + 1:]
date_pattern = re.compile(r"\d{4}-\d{2}-\d{2}")
clean_lines = [line for line in data_lines if date_pattern.match(line.split()[0])]

# Create DataFrame
data_str = '\n'.join(clean_lines)
df = pd.read_csv(StringIO(data_str), sep=r'\s+', names=["DATE", "TIME", "DOY", "DST"])
df['datetime'] = pd.to_datetime(df['DATE'] + ' ' + df['TIME'])

# Plot with storm thresholds
plt.figure(figsize=(12, 6))
plt.plot(df['datetime'], df['DST'], label='Dst Index', color='violet')

# Storm thresholds
thresholds = {
    "Disturbed": -30,
    "Moderate storm": -50,
    "Intense storm": -100,
    "Superstorm": -250
}

for label, value in thresholds.items():
    plt.axhline(value, color='indigo', linestyle='--', linewidth=0.8)
    plt.text(df['datetime'].iloc[-1], value + 2, label, color='indigo', fontsize=8, va='bottom', ha='right')

plt.title('Dst Index with Geomagnetic Storm Thresholds')
plt.xlabel('Date')
plt.ylabel('Dst (nT)')
plt.grid(True)
plt.tight_layout()
plt.show()

```

## Part 3: Tsunami Plot

```python
# Define parameters
h_values = np.linspace(50, 5000, 100)  # Ocean depths from 50m to 5000m
c_values = [50, 150, 250]  # Different tsunami gravity wave speeds in m/s

# Constants
n = 1  # Tsunami wave height (arbitrary for this case)
A = 1  # Complex scaling factor, ignored for this case

# Calculate the induced field ratio b_z / B_z for different speeds
plt.figure(figsize=(8, 6))

for c in c_values:
    b_z_B_z = (n / h_values) * c * A 
    plt.plot(h_values, b_z_B_z, label=f'c = {c} m/s')

# Labeling
plt.title('Normalized Induced Magnetic Field vs Ocean Depth for Different Tsunami Speeds')
plt.xlabel('Ocean Depth (m)')
plt.ylabel('Normalized Induced Magnetic Field (b_z / B_z)')
plt.legend(title="Tsunami Speed")
plt.grid(True)

plt.show()
```
